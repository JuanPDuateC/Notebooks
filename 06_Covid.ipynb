```python
# covid por departamento
## de un solo nivel cd = covid.groupby(["departamento_nom"])["edad"].describe()
## de dos niveles
cd = covid.groupby(["Nombre departamento"])[["Edad"]].describe()
cd
```

```python
cd.shape
```

```python
cd.columns
```

```python
# Eliminar el nivel
cd.columns = cd.columns.droplevel(0)
```

```python
#contar casos por departamento
cd["count"]
```

```python
# graficar la cantidad de casos por departamento
fig, ax = plt.subplots(figsize=(16,6))
# alpha cambia la intensidad del color
# X es una de las columnas en este caso ya es index por la organizacion del group by  y la Y seria la cantidad representada en height
ax.bar(x=cd.index, height=cd["count"], color="crimson", alpha=0.86)
# https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text
plt.xticks(rotation=45, ha="right" )
plt.show
```

```python
fig, ax = plt.subplots(figsize=(16,6))
# alpha cambia la intensidad del color
# X es una de las columnas en este caso ya es index por la organizacion del group by  y la Y seria la cantidad representada en height
ax.bar(x=cd.index, height=cd["mean"], color="crimson", alpha=0.86)
# https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text
plt.xticks(rotation=45, ha="right" )
plt.show
```

```python
covid.columns
```

```python
# modificar las tipos de datos e object a  fechas

for fecha in ['fecha reporte web', 'Fecha de notificación', 'Fecha de inicio de síntomas', 'Fecha de muerte', 'Fecha de diagnóstico', 'Fecha de recuperación']:
    covid[fecha]= pd.to_datetime(covid[fecha], format="%d/%m/%Y %H:%M:%S")
```

```python
covid.info()
```

```python
cf = covid.groupby(["Fecha de diagnóstico"])[["Edad"]].describe()
cf.columns = cf.columns.droplevel(0)
cf
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.set(title="cantidad de pacientes por dia", xlabel="Fechas", ylabel="Numero de pacientes")
ax.bar(x=cf.index, height=cf["count"], color="crimson", alpha=0.86)
# https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text
plt.xticks(rotation=45, ha="right" )
plt.show
```

```python
cf["acumulados"] = cf["count"].cumsum()
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.plot(cf.index, cf["acumulados"], "g:")
ax.set(title="cantidad de pacientes por dia", xlabel="Fechas", ylabel="Numero de pacientes")
ax.bar(x=cf.index, height=cf["acumulados"], color="crimson", alpha=0.86)
# https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text
# plt.xticks(rotation=45, ha="right" )
ax.grid()
plt.show
```
